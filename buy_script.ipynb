```python
import time
import re
import traceback
from selenium import webdriver
from selenium.webdriver.support.ui import Select,WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys as keys
import schedule
from datetime import timedelta
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary

#################################
#単勝馬券　購入スクリプト
#################################


def buy(race_id,id,pw,IPAT_password,horse_num,bat):
    
    
    #netkeiba.comログイン
    #####################################
    options = Options()

    profile_path = "C:/sers/ユーザー名/AppData/Local/Google/Chrome/User Data/Default"
    options.add_argument('--user-data-dir=' + profile_path)
    Chromeexe = "chromedriver.exe"

    options.add_argument('--disable-gpu')
    #options.add_argument('--headless')    # ヘッドレスモードに
    driver = webdriver.Chrome(executable_path=Chromeexe,options=options)
    wait = WebDriverWait(driver,10)






    URL = "https://regist.netkeiba.com/account/?pid=login&action=auth"

    driver.get(URL)
    time.sleep(1)
    wait.until(EC.presence_of_all_elements_located)



    # ID/PASSを入力
    login_id = driver.find_element_by_name("login_id")
    login_id.send_keys(id)
    password = driver.find_element_by_name("pswd")
    password.send_keys(pw)
    password.send_keys(keys.ENTER)


    wait.until(EC.presence_of_all_elements_located)
    #####################################
    
    
    
    
    #購入処理
    buy_url = 'https://race.netkeiba.com/ipat/dispatch.html?race_id={0}'.format(race_id)

    driver.get(buy_url)

    time.sleep(2)

    password = IPAT_password
    wait.until(EC.presence_of_all_elements_located)

    # PAT購入画面に遷移・ログイン
    time.sleep(2)
    #馬単選ぶ
    driver.find_element_by_xpath('//*[@id="tag-ipat"]/ipat/div/section[2]/div/fieldset/dl/dd[1]/a').click()
    #馬番選ぶ
    driver.find_element_by_xpath('/html/body/div[1]/div/ipat/div/section[3]/div/section[1]/div/table/tbody/tr[{0}]/td[5]'.format(horse_num)).click()
    #賭け金入力
    driver.find_element_by_xpath('/html/body/div[1]/div/ipat/div/section[3]/div/section[2]/div[1]/div[2]/div[2]/div[1]/div[1]/input').send_keys(bat)
    #追加
    driver.find_element_by_xpath('/html/body/div[1]/div/ipat/div/section[3]/div/section[2]/div[1]/div[2]/div[2]/div[1]/div[2]/button').click()
    time.sleep(3)
    wait.until(EC.presence_of_all_elements_located)
    #送信
    driver.find_element_by_xpath('/html/body/div[1]/div/ipat/div/section[3]/div/section[2]/div[4]/button').click()

    wait.until(EC.presence_of_all_elements_located)

    time.sleep(3)
    
    # 新しいタブに切り替える
    driver.switch_to.window(driver.window_handles[1])
    time.sleep(3)
    #driver.switch_to.frame(driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div[2]/div[1]/iframe'))


    #パスワード入力
    
    driver.find_element_by_xpath('/html/body/div/div/jraipat/div[1]/div/form/div/dl[2]/dd/input').send_keys(password)
    time.sleep(1)
    #ログイン
    driver.find_element_by_xpath('/html/body/div/div/jraipat/div[1]/div/form/div/div[2]/a').click()

    wait.until(EC.presence_of_all_elements_located)
    time.sleep(3)

    #金額確認入力
    driver.find_element_by_xpath('/html/body/div[2]/div/div/dl/dd/input').send_keys(bat*100)
    time.sleep(10)
    #送信
    driver.find_element_by_xpath('/html/body/div[2]/div/div/div[2]/ul/li[2]/a').click()
    time.sleep(3)
    wait.until(EC.presence_of_all_elements_located)
    Alert(driver).accept() #alertやconfirmのOKを押す。
    time.sleep(1)
    wait.until(EC.presence_of_all_elements_located)
    #driver.switch_to.default_content()
    time.sleep(1)
    #閉じる
    #driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div[2]/button[4]').click()

    wait.until(EC.presence_of_all_elements_located)

    time.sleep(1)
    

    # 前タブに切り替える
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(3)
    #馬券投票データ保存
    driver.find_element_by_xpath('//html/body/div[5]/div/div[2]/ul/li[1]/a/input').click()
    time.sleep(1)

    print('購入完了\nrace_id:{0}\n馬番:{1}\n金額:{2}'.format(race_id,horse_num,bat*100))
    driver.quit()
    
    
def main():
    netkeiba_id = "" #netkeiba アカウントID
    netkeiba_pw = "" #netkeiba パスワード
    IPAT_pw = "" #IPAT パスワード
    
    print("Race ID 入力")
    race_id = input()
    print("賭ける馬番")
    horse_num = int(input()) #賭ける馬番
    print("賭ける金額")
    bat = int(input())//100
    buy(race_id,netkeiba_id,netkeiba_pw,IPAT_pw,horse_num,bat)


if __name__ == "__main__":
    main()
```
